```python
#load dependencies
from chatterbot import ChatBot
from chatterbot.trainers import ChatterBotCorpusTrainer
import pandas as pd
```

```python
#make language class compatible with spacy
class ENGSM:
    ISO_639_1 = "en_core_web_sm"
```

```python
#create bot
chatbot = ChatBot("Obvious", tagger_language=ENGSM)

trainer = ChatterBotCorpusTrainer(chatbot)
#train bot
trainer.train("chatterbot.corpus.english")

print(chatbot.get_response("Hello, how are you today?"))
```

```python
#new df to have less time is a factor
df = questions[0:10000]
df.head()
```

```python
#create new bot for trivia
#create bot
trivia = ChatBot("Trivia", tagger_language=ENGSM)

from chatterbot.trainers import ListTrainer

triviaTrain = ListTrainer(trivia)
#train bot
for i in range(len(df)):
    triviaTrain.train([df[" Question"][i], df[ ' Answer'][i]])
```

```python
#randomly shuffle df
df = df.sample(frac=1)
df.head()
```

```python
#create needed lists fro final df
chatAnswer =[]
triviaAnswer = []
answer = []
category = []
question = []
value = []
```

```python
df = df.reset_index()
df.head()
```

```python
#only using first 500 for time
for i in range(500):
    value.append(df[" Value"][i])
    answer.append(df[" Answer"][i])
    category.append(df[" Category"][i])
    question.append(df[" Question"][i])
    chatAnswer.append(chatbot.get_response(df[" Question"][i]))
    triviaAnswer.append(trivia.get_response(df[" Question"][i]))
    
    
```

```python
#creating final df
dict = {"Question":question, "Value":value,"Category":category,"Answer":answer,"ChatBot":chatAnswer,"TriviaBot":triviaAnswer}

final = pd.DataFrame(dict)
final.head()
```

```python
#comparing accuracy

botac = 0
triviaac = 0

for i in range(len(final)):
    if(str(final["ChatBot"][i]) in str(final["Answer"][i])):
        botac +=1
        
    if(str(final["TriviaBot"][i]) in str(final["Answer"][i])):
        triviaac +=1
print(botac)
print(triviaac)
```
